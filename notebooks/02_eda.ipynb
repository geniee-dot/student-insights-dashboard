```python
import pandas as pd      # for tables like Excel
import matplotlib.pyplot as plt  # for charts
import seaborn as sns    # nicer charts than matplotlib

df = pd.read_csv('../data/processed/cleaned_student_data.csv')
df.head()

df.info()
df.describe()


plt.figure(figsize=(8,5))
sns.histplot(df['final_grade'], bins=15, kde=True)
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.show()


```

# Step 4 â€” Exploratory Data Analysis
Here we explore grade distributions, absences, study time, and identify at-risk students.


```python
# Create Pass/Fail column
df['performance_status'] = df['final_grade'].apply(lambda x: 'Pass' if x >= 10 else 'Fail')

# Check distribution
df['performance_status'].value_counts()
```

```python
pass_rate = (df['performance_status'] == 'Pass').mean() * 100
print(f"Pass Rate: {pass_rate:.2f}%")
```
